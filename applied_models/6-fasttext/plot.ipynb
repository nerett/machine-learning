```python
import re
```

```python
pattern = r'Epoch \[(\d+)/\d+\], Step \[(\d+)/\d+\], Loss: (\d+\.\d+)'
re.findall(pattern, "Epoch [123/5], Step [456/2345], Loss: 9.1993, Perplexity: 9890.2299")
```

```python
losses = []
for fname in ['2.log', '4.log']:
    with open('./' + fname, 'r') as f:
        for l in f:
            if (loss_match := re.search(r'Loss: (\d+\.\d+)', l)):
                losses.append(float(loss_match.group(1)))
```
